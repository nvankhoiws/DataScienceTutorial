```python
import pandas as pd
all_ages = pd.read_csv('all-ages.csv')
all_ages.head(5)
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.head(5)
```

```python
# Summarizing Major Categories
aa_cat_counts = dict()
rg_cat_counts = dict()

# print(all_ages['Major_category'].unique())
# print(type(all_ages['Major_category'].unique()))

import numpy

print('===========Method 1: Use loop====================\n')
# Method 1: Use loop
def get_total_for_cat(df):
    unique_cat = df['Major_category'].unique()
    temp_dict = dict()
    for cat in unique_cat:
        cat_bool = df['Major_category'] == cat
        cat_bool_val = df[cat_bool]
        sum_val = cat_bool_val['Total'].sum()
        temp_dict[cat] = sum_val
    return temp_dict
aa_cat_counts = get_total_for_cat(all_ages)
rg_cat_counts = get_total_for_cat(recent_grads)
print(aa_cat_counts)
print('\n')
print(rg_cat_counts)

print('===========Method 2: Use pivot_table====================\n')
# Method 2: Use pivot_table
def get_total_for_cat(df):
    unique_cat = df['Major_category'].unique()
    group_by_cat = df.pivot_table(index='Major_category',values='Total',aggfunc=numpy.sum)
    temp_dict = dict()
    for cat in unique_cat:
        temp_dict[cat] = group_by_cat.loc[cat, 'Total']
    return temp_dict

aa_cat_counts = get_total_for_cat(all_ages)
rg_cat_counts = get_total_for_cat(recent_grads)
print(aa_cat_counts)
print('\n')
print(rg_cat_counts)
```
